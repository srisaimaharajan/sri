```python
df['Total Sales'] = df[['Q1 2019','Q2 2019','Q3 2019','Q4 2019','Q1 2020','Q2 2020','Q3 2020','Q4 2020']].sum(axis = 1)
```

```python
#5 How can you calculate the total sales for each brand across all quarters?
total_sale=df.groupby('Brand').agg({'Total Sales':['sum']})
total_sale
```

```python
df['Q1 Sales'] = df['Q1 2019'] + df['Q1 2020']
```

```python
#6 How can you filter the DataFrame to show only the rows where the Q1 Sales are greater than 10,000?
greater = df[df['Q1 Sales']> 10000][['Brand','Q1 Sales']]
greater
```

```python
#7.How can you group the data by “AUTOGROUP” and calculate the total sales for each group?
df.head(2)
total_sale=df.groupby('Autogroup').agg({'Total Sales':['sum']})
total_sale
```

```python
#8.How can you find the brand with the highest total sale?
high= df.loc[df['Total Sales'].idxmax()][['Brand','Total Sales']]
high.to_frame()
```

```python
high = df.iloc[df['Total Sales'].idxmax()][['Brand','Total Sales']]
high.to_frame()
```

```python
top_brand=df[df['Total Sales'] == df['Total Sales'].max()][['Brand','Total Sales']]
top_brand
```

```python
#9.How can you add a new column that calculates the average quarterly sales for each brand?
df['AVG_Q_SALE1']=df[['Q1 2019','Q2 2019','Q3 2019','Q4 2019']].mean(axis = 1)
df
```

```python
df['AVG_Q_SALE']=df[['Q1 2020','Q2 2020','Q3 2020','Q4 2020']].mean(axis = 1)
df
```

```python
#10 How can you sort the DataFrame by total sales in descending order?
df.sort_values(by='Total Sales',ascending=False)[['Brand','Total Sales']]
```

```python
#11 How can you find the average sales for each quarter across all brands?
avg_sale=df[['Q1 2019','Q2 2019','Q3 2019','Q4 2019','Q1 2020','Q2 2020','Q3 2020','Q4 2020']].mean()
avg_sale.to_frame()
```

```python
#12 How can you find the total sales for each autogroup?
auto_sale=df.groupby('Autogroup')['Total Sales'].agg('sum')
auto_sale.to_frame()
```

```python
#13 How can you plot the total sales for each brand using a bar chart?
total_sales_by_brand=df.groupby('Brand')['Total Sales'].sum()
total_sales_by_brand.plot(kind='bar', figsize = (15, 4))
```

```python
#14 How can you calculate the percentage contribution of each quarter to the total sales for each brand?
df['Q1 sales %'] = df['Q1 2019'] / df['Total Sales'] * 100
df
```

```python
df['Q2 sales %'] = df['Q2 2019'] / df['Total Sales'] * 100
df['Q3 sales %'] = df['Q3 2019'] / df['Total Sales'] * 100
df['Q4 sales %'] = df['Q4 2019'] / df['Total Sales'] * 100
df
```

```python
#15 How can you create a pivot table to show the total sales for each brand per quarter?



pd.pivot_table(data = df, values = ['Q1 2020', 'Q2 2020', 'Q3 2020', 'Q4 2020'], 
               index = 'Brand', aggfunc = 'sum',
)
```

```python
#16 How can you rename the columns Q1Sales, Q2Sales, Q3Sales, and Q4Sales to Quarter1, Quarter2,Quarter3,and Quarter4
df.rename(columns={'Q1 2019': 'Quarter 1',
                   'Q2 2019': 'Quarter 2',
                   'Q3 2019': 'Quarter 3',
                   'Q4 2019': 'Quarter 4',
                   'Q1 2020':'Quarter1 2020',
                   'Q2 2020':'Quarter2 2020',
                   'Q3 2020':'Quarter3 2020',
                   'Q4 2020':'Quarter4 2020',
                  }, inplace=True)
df
```
