```python
df.rename(columns = {'Province/State': 'state', 
                     'Country/Region':'country'}, inplace = True)
df
```

```python
##2.convert the date column into date time
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
##4.cal the mortality rate for each country
df['mortality']= df['Deaths'] / df['Confirmed']*100
mortality_rate =df.groupby('country')['mortality'].agg('count')
mortality_rate.to_frame() 
```

```python
##5.CAL RECOVERY RATE FOR EACH COUNTRY
df['recovery'] = df['Recovered']/df['Confirmed']*100
recovery_rate = df.groupby('country')['recovery'].agg('count')
recovery_rate.to_frame()
```

```python
##6.find the country with the highest average mortality rate
df['mortality']= df['Deaths'] / df['Confirmed']*100
avg_mortality_rate =df.groupby('country')['mortality'].agg('mean')
avg_mortality_rate.to_frame() 
```

```python
highest_mortality_rate = avg_mortality_rate.idxmax()
highest_mortality_rate 
```

```python
##7.find the country with the highest average recovery rate
df['recovery'] = df['Recovered']/df['Confirmed']*100
recovery_rate = df.groupby('country')['recovery'].agg('mean')
recovery_rate.to_frame()
```

```python
highest_recovery_rate = recovery_rate.idxmax()
highest_recovery_rate
```

```python
df['month']= df['Date'].dt.month
df
```

```python
df['year'] = df['Date'].dt.year
df
```

```python
df['day'] = df['Date'].dt.day
df
```

```python
##8.calculate the total confirmed,deaths,recovered, and active cases per month
monthly_cases=df.groupby('month')[['Confirmed','Deaths','Recovered','Active']].agg('sum')
monthly_cases
```

```python
##9.calculate the monthly new confirmed cases world wide
new_confirmed_cases=df.groupby('month')['Confirmed'].sum().diff().fillna(0)
new_confirmed_cases.to_frame()
```

```python
##10.calculate the monthly new deaths cases world wide
new_death_cases=df.groupby('month')['Deaths'].sum().diff().fillna(0)
new_death_cases.to_frame()
```

```python
##11.calculate the monthly new recovered cases world wide
new_recovered_cases=df.groupby('month')['Recovered'].sum().diff().fillna(0)
new_recovered_cases.to_frame()
```

```python
##12.calculate the monthly new active cases world wide
new_active_cases=df.groupby('month')['Active'].sum().diff().fillna(0)
new_active_cases.to_frame()
```

```python
##13.top 5 countries with highest number of deaths on a specific date
specific_date = '2020-07-27'
highest_no_of_deaths=df[df['Date']==specific_date].nlargest(5, 'Deaths')[['Date','country','Deaths']]
highest_no_of_deaths
```

```python
##14.CALCULATE THE AVERAGE DAILY INCREASE IN CONFIRMED CASES FOR EACH COUNTRY
daily_confirmed_cases=df.groupby(['day','country'])['Confirmed'].mean()
daily_confirmed_cases.to_frame()
```

```python
##15.CALCULATE THE AVERAGE DAILY INCREASE IN deaths CASES FOR EACH COUNTRY
daily_deaths_cases=df.groupby(['day','country'])['Deaths'].mean()
daily_deaths_cases.to_frame()
```

```python
##16 CALCULATE THE AVERAGE DAILY INCREASE IN recovered CASES FOR EACH COUNTRY
daily_recovered_cases=df.groupby(['day','country'])['Recovered'].mean()
daily_recovered_cases.to_frame()
```

```python
##17.find the date with the highest number of new confirmed cases world wide
new_confirmed_cases=df.groupby('Date')['Confirmed'].sum().diff().fillna(0)
new_confirmed_cases.to_frame()
```

```python
highest_confirmed_cases=new_confirmed_cases.idxmax()
highest_confirmed_cases
```

```python
##18.find the date with the highest number of new death cases world wide
new_death_cases=df.groupby('Date')['Deaths'].sum().diff().fillna(0)
new_death_cases.idxmax()
```

```python
##19.find the date with the highest number of new recovered cases world wide
new_recovered_cases=df.groupby('Date')['Recovered'].sum().diff().fillna(0)
new_recovered_cases.idxmax()
```

```python
##20.CALCULATE THE CUMULATIVE CONFIRMED CASES PER COUNTRY OVER TIME
cumulative_confirmed_cases=df.groupby(['country','Date'])['Confirmed'].sum().groupby(level=0).cumsum()
cumulative_confirmed_cases.to_frame()
```
